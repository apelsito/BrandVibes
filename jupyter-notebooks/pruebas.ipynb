# Generos escuchados

```python
from collections import Counter

# Obtener los artistas más escuchados
top_artists = sp.current_user_top_artists(limit=50, time_range='short_term')  # Cambia el 'time_range' según prefieras

# Extraer los géneros de los artistas
genres = []
for artist in top_artists['items']:
    genres.extend(artist['genres'])  # Agrega todos los géneros de cada artista

# Contar las ocurrencias de cada género
genre_count = Counter(genres)

# Ordenar y mostrar el top 10 géneros más escuchados
print("🎼 Top de géneros más escuchados:")
for genre, count in genre_count.most_common(40):
    print(f"{genre}: {count} artistas asociados")
```

# Agrupar géneros por categorías generales

```python
# Mapear subgéneros a géneros principales
general_genres = {
    "Pop": ["pop", "pop dance", "dance pop", "latin pop", "colombian pop", "pop venezolano", "uk pop"],
    "Electrónica": ["edm", "electro house", "big room", "progressive electro house", "dutch house", "slap house",
                    "house", "brostep", "trance", "melbourne bounce", "melbourne bounce international",
                    "canadian electronic", "uk dance", "covertronica", "eurobeat", "super eurobeat",
                    "progressive house", "complextro"],
    "Urbano": ["urbano latino", "reggaeton", "trap latino", "reggaeton colombiano", "urbano espanol", "argentine hip hop",
               "trap argentino", "drill espanol"],
    "Rock": ["rock", "classic rock", "psychedelic rock", "british invasion", "merseybeat"],
    "Mandopop": ["mandopop"],
    # Agrega más géneros principales si es necesario
}

generalized_counts = {}

# Recorrer el conteo de géneros específicos
for genre, count in genre_count.items():
    # Buscar a qué género principal pertenece
    for general_genre, subgenres in general_genres.items():
        if genre in subgenres:
            generalized_counts[general_genre] = generalized_counts.get(general_genre, 0) + count
            break

# Ordenar por conteo descendente
generalized_counts = dict(sorted(generalized_counts.items(), key=lambda x: x[1], reverse=True))

# Mostrar los resultados
print("🎼 Géneros generales más escuchados:")
for general_genre, count in generalized_counts.items():
    print(f"{general_genre}: {count} artistas asociados")
```
